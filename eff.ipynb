## RSA Encryption Efficiency Test

```python
times=[]
sizes=[]
message="lets encrypt this message"

for i in range(16,1024,4):
    public, private = RSA.generate_keys(i)
    t1 = time.time()
    # encode our message
    encoded_messages = RSA.en.encode(message)
    for encoded in encoded_messages:
        enc = RSA.encrypt(int(encoded),public)

    t2 = time.time()
    sizes.append(i)
    times.append(t2-t1)

plt.plot(sizes, times)
plt.xlabel("Key size)")
plt.ylabel("Time_taken (s)")
plt.title("Encryption Efficiency (timing)")
plt.show()
```
